```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Specify image name (replace with the uploaded file's name)
image_path = "proflileME.png"  # Replace with your image file name
```

```python
# Read the image
img = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Convert grayscale image to float32 (required for Harris)
gray_float = np.float32(gray)

# Apply the Harris Corner Detector
harris_corners = cv2.cornerHarris(src=gray_float, blockSize=2, ksize=3, k=0.04)

# Dilate the corners to enhance visibility
harris_corners = cv2.dilate(harris_corners, None)
```

```python
# Create a copy of the original image to display corners
img_with_corners = img.copy()

# Threshold for an optimal value (this depends on the image and task)
threshold = 0.01 * harris_corners.max()

# Mark corners in red
img_with_corners[harris_corners > threshold] = [0, 0, 255]
```

```python
plt.figure(figsize=(12, 8))

# Original Image
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

# Grayscale Image
plt.subplot(1, 3, 2)
plt.title("Grayscale Image")
plt.imshow(gray, cmap='gray')
plt.axis("off")

# Image with Corners
plt.subplot(1, 3, 3)
plt.title("Image with Corners")
plt.imshow(cv2.cvtColor(img_with_corners, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
cv2.imwrite("Harris_Corners.jpg", img_with_corners)

from google.colab import files
files.download("Harris_Corners.jpg")
```
